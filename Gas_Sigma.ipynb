```python
def columns_graph(df):
    for i in df.keys():
        plt.figure(figsize=(18,8))
        plt.plot(df[i])
        plt.title(i)
        plt.show()
```

```python
def outlier_graph(df):
    for i in df.keys():
        fig,ax = plt.subplots(figsize=(18,3))
        plt.subplot(1,3,1)
        sns.boxplot(df[i])
        plt.subplot(1,3,2)
        sns.scatterplot(df[i])
        plt.subplot(1,3,3)
        sns.histplot(df[i], kde=True)
        plt.title(i)
        plt.show()
```

```python
def calculate_sigma_values(df):
    
    mean_value = np.mean(df)
    median_value = np.median(df)
    min_value = np.min(df)
    max_value = np.max(df)
    
    std_dev = np.std(df)
    sigma_3_neg = (mean_value - 3 * std_dev)
    sigma_2_neg = (mean_value - 2 * std_dev)
    sigma_neg = (mean_value - std_dev)
    sigma = (mean_value + std_dev)
    sigma_2 = (mean_value + 2 * std_dev)
    sigma_3 = (mean_value + 3 * std_dev)
        
    return min_value,sigma_3_neg,sigma_2_neg,sigma_neg,mean_value,sigma,sigma_2,sigma_3,max_value,median_value,std_dev
```

```python
for i in df2.keys():
    sigma = calculate_sigma_values(df2[i])
    print(i, sigma)
```
